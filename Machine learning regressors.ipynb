## Machine learning

```python
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
from sklearn.model_selection import train_test_split
```

### Linear Regression

```python
house.columns
```

```python
X = house[['bedrooms', 'bathrooms', 'sqft_living',
       'sqft_lot', 'floors', 'waterfront', 'view', 'condition', 'grade',
       'sqft_above', 'sqft_basement', 'yr_built', 'yr_renovated', 'zipcode',
       'lat', 'long', 'sqft_living15', 'sqft_lot15']]
```

```python
y = house['price']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=52)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
print("Mean squared error is:", mean_squared_error(y_test, y_pred))
print("Mean absolute error is:", mean_absolute_error(y_test, y_pred))
print("R2 score is:", r2_score(y_test, y_pred))
print("Root mean square:", np.sqrt(mean_squared_error(y_test, y_pred)))
```

```python
from sklearn.metrics import accuracy_score
```

```python
r2_score(y_test, y_pred)
```

```python
mean_absolute_error(y_test, y_pred)
```

```python
from sklearn.metrics import classification_report, confusion_matrix # We use this for data which is not continuous
```

```python
print("mean Squared error is:", mean_squared_error(y_test, y_pred))
```

```python
model.coef_
```

```python
coefficients = pd.DataFrame(model.coef_, X.columns, columns = ["Coefficient values"])
coefficients
```

```python
plt.scatter(y_test, y_pred)
```

## Decision Tree

```python
dtr = DecisionTreeRegressor()
```

```python
dtr.fit(X_train, y_train)
```

```python
y_pred = dtr.predict(X_test)
```

```python
print("Mean absolute error is:", mean_absolute_error(y_test, y_pred))
print("Mean squared error is:", mean_squared_error(y_test, y_pred))
print("Root mean square is (RMSE):", np.sqrt(mean_squared_error(y_test, y_pred)))
print("R2_score is:", r2_score(y_test, y_pred))
```

Decision tree regressor gives a smaller RMSE than linear regression

```python
plt.scatter(y_test, y_pred)
```

```python
coefficients = pd.DataFrame(dtr.coef_, X.columns, columns = ["Coefficients"])
coeficients
```

## Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rfr = RandomForestRegressor()
```

```python
rfr.fit(X_train, y_train)
```

```python
y_pred = rfr.predict(X_test)
y_pred
```

```python
plt.scatter(y_test, y_pred)
plt.title("Scatterplot")
plt.xlabel("y_test")
plt.ylabel("y_pred")
```

```python
print("Mean_absolute error is:", mean_absolute_error(y_test, y_pred))
print("Mean squared error is:", mean_squared_error(y_test, y_pred))
print("Root mean squared error is:", np.sqrt(mean_squared_error(y_test, y_pred)))
print("R2 score is:", r2_score(y_test, y_pred))
```

Random  forest regressor is better than both decision tree regressor and linear regression
